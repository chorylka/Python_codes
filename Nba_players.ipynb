```python
#OurOwnFunction
#This functions makes the same as above but now using the plot as a function passing a list of players as
#parameter
```

```python
def myplot(playerlist):
    for name in playerlist:
        plt.plot(Salary[Pdict[name]], c='Black', ls='--', marker='s', ms=7, label=name)    
    
    plt.legend(loc='upper left', bbox_to_anchor=(1,1))
    plt.xticks(list(range(0,10)), Seasons, rotation='vertical')
    plt.show()
```

```python
myplot(["KobeBryant", "JoeJohnson", "ChrisBosh"])
```

```python
#Advanced Function Design


def myplot(playerlist):
    Col = {"KobeBryant":"Black","JoeJohnson":"Red","CarmeloAnthony":"Green"}
    for name in playerlist:
        plt.plot(Salary[Pdict[name]], c=Col[name], ls='--', marker='s', ms=7, label=name)    
    
    plt.legend(loc='upper left', bbox_to_anchor=(1,1))
    plt.xticks(list(range(0,10)), Seasons, rotation='vertical')
    plt.show()
    
    
```

```python
#Fixing up the inputs:
def myplot(data, playerlist = Players):
    Col = {"KobeBryant":"Black","JoeJohnson":"Red","CarmeloAnthony":"Green"}
    Mrk = {"KobeBryant":"s","JoeJohnson":"o","CarmeloAnthony":"d"}
    for name in playerlist:
        plt.plot(data[Pdict[name]], c=Col[name], ls='--', marker=Mrk[name], ms=7, label=name)    
    
    plt.legend(loc='upper left', bbox_to_anchor=(1,1))
    plt.xticks(list(range(0,10)), Seasons, rotation='vertical')
    plt.show()
    
```
